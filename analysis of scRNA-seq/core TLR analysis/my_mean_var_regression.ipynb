```python
#print(induced_data.head)

genelist = induced_data.iloc[:,0].reset_index(drop=True).drop_duplicates().tolist()
#genelist += [genelist.pop(24)] #manually moving index 24 TNF to end so visible on plots
#genelist = []
#with open("eq3conditions_lpspicUpreg.txt","r") as genefile:
#    for g in genefile:
#        genelist.append(g.rstrip())
#genelist = ["IL1B", "IL1A"]
i=0

plt.subplots(1,1,figsize=(5,4.5))
plt.title("Combined Linear Regression")
axes = plt.gca()
axes.set_xlim([0,3000])
axes.set_ylim([0,25000000])

with open("summary_mean_var_812_no_zeros.csv","a") as logfile:
        logfile.write("%s,%s,%s,%s,%s,%s,%s,\n" % ('gene','N','intercept','slope','min x','max x','R2'))
  
for test_gene in genelist:
    i+=1
    print(i)
    summary_stats = induced_data[induced_data["Gene"]==test_gene]
    #print(summary_stats)
    if (len(summary_stats.loc[summary_stats['Mean'] == 0,['Mean']])>0) :
        print("Missing data for %s" % test_gene)
        with open("errorlog_812.csv","a") as logfile:
            logfile.write("%s\n" % test_gene)
        continue
    
    print("Working on %s mean/var regression..." % test_gene)
    lm = linear_model.LinearRegression()
    model = lm.fit(summary_stats['Mean'].values.reshape(-1, 1),summary_stats['Var'])
    print("Intercept:\t",lm.intercept_)
    print("Coefficients:\t",lm.coef_)
    scoreAll = lm.score(summary_stats['Mean'].values.reshape(-1, 1),summary_stats['Var'])
    print("R^2 score:\t",scoreAll)

    linecol = 'gray'

    """
    if(test_gene == 'TNF') :
        linecol = 'red'
    elif(test_gene == 'IL1A'):
        linecol = 'blue'
    elif(test_gene == 'IL1B'):
        linecol = 'green'
    elif(test_gene == 'CCL4'):
        linecol = 'orange'
    elif(test_gene == 'CCL17'):
        linecol = 'purple'
    """
    """    if test_gene in grp:
        
        if test_gene in grp2:
            linecol = 'blue'
        elif test_gene in grp1:
            linecol = 'red'"""
    if (max(summary_stats['Mean'])>100):

        plt.plot(summary_stats['Mean'], lm.predict(summary_stats['Mean'].values.reshape(-1,1)), color=linecol, linewidth=1)

        with open("summary_mean_var_812_no_zeros.csv","a") as logfile:
            logfile.write("%s,%s,%s,%s,%s,%s,%s,\n" % (test_gene,len(summary_stats['Mean']),lm.intercept_,float(lm.coef_),min(summary_stats['Mean']), max(summary_stats['Mean']), scoreAll))
            print("%s,%s\n" % (test_gene,scoreAll))

plt.ylabel("Var")
plt.xlabel("Mean")
#plt.savefig("global_regressions/mean_var_composite.png")
plt.savefig("my_mean_var_composite_grps_no_zeros.png")
plt.close()


```
