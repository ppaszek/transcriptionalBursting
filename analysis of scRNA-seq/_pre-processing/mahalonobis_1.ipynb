```python
from scipy.spatial.distance import mahalanobis 
import scipy as sp 
import pandas as pd 
from scipy.stats import chi2 
def mahalanobisR(X,meanCol,IC):
    m = []
    for i in range(X.shape[0]):
        m.append((X.iloc[i,:].name,mahalanobis(X.iloc[i,:],meanCol,IC) ** 2))
    return(m) 

 #x = pd.read_csv('induced_genes_processed_data_all.csv') #read in to dataframe 
    
x = pd.read_csv('induced_genes_outliers_removed_all_812_3runs.csv') #read in to dataframe 

x = x.iloc[:,1:] #remove duplicate id col
#print(x)
#genelist = ['IL1B','TNF','CCL4','CCL17','IL1A'] 
genelist = x.iloc[:,0].reset_index(drop=True).drop_duplicates().tolist()

x_output = pd.DataFrame(columns=x.columns) 
m_dist_threshold = chi2.ppf(0.95,2) 

for test_gene in genelist:
    x_orig = x.loc[(x['Gene'] == test_gene)&(x['Len'] > 10),["Mean","Var"]]
    x_filtered = pd.DataFrame(columns=x_orig.columns)
    x_final = pd.DataFrame(columns=x.columns)
    Sx_orig = x_orig.cov().values
    Sx_orig = sp.linalg.inv(Sx_orig)
    mean_orig = x_orig.mean().values
    outliers_0 = []
    mR0 = mahalanobisR(x_orig,mean_orig,Sx_orig)
    for mRi in mR0:
        if mRi[1] > m_dist_threshold:
            outliers_0.append(x.loc[mRi[0],["Condition"]].tolist()[0])
        else :
            x_filtered.loc[mRi[0]] = x_orig.loc[mRi[0]]
    Sx_filtered = x_filtered.cov().values
    Sx_filtered = sp.linalg.inv(Sx_filtered)
    mean_filtered = x_filtered.mean().values
    outliers_1 = []
    mR1 = mahalanobisR(x_filtered,mean_filtered,Sx_filtered)
    for mRi in mR1:
        if mRi[1] > m_dist_threshold:
            outliers_1.append(x.loc[mRi[0],["Condition"]].tolist()[0])
        else :
            x_final.loc[mRi[0]] = x.loc[mRi[0]]
    lps_count = len(x_final[x_final['Condition'].str.contains('LPS')])
    pam_count = len(x_final[x_final['Condition'].str.contains('PAM')])
    pic_count = len(x_final[x_final['Condition'].str.contains('PIC')])
    total_count = len(x_final[x_final['Condition'].str.contains('')])
    print("%s\t%s\t%s\t%s\t%s\t%s\t%s" % (test_gene,total_count,lps_count,pam_count,pic_count, ";".join(outliers_0),";".join(outliers_1)))
    with open("outlier_summary.csv","a") as logfile:
        logfile.write("%s,%s,%s,%s,%s,%s,%s\n" % (test_gene,total_count,lps_count,pam_count,pic_count, ";".join(outliers_0),";".join(outliers_1)))
    x_output = x_output.append(x_final) 
x_output.to_csv('induced_genes_outliers_removed_all_812_4runs.csv')

```
